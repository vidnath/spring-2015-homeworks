According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# Write your code here
snow_group = snow_data.groupby('DATE') # easy collection of dates
snow_days = dict()  # will be of the form {date: snowfall_amount}

for day, group in snow_group:
    date = day.date()
    snow_days[date] = group.SNOW.max() / 10.  # convert to cm

# inspiration: http://stackoverflow.com/questions/18837262/convert-python-dict-into-a-dataframe
snow = pd.Series(snow_days,name='Amount')  # convert into a series to plot correctly
snow.index.name = 'Date'

snow_plot = snow.plot(title = 'Snowfall in Boston Area')
snow_plot.set_ylabel('Amount (cm)')
print snow.head(5)
````
