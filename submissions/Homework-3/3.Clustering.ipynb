## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

```python
#boston.info()
#print len(boston.LATITUDE)
nbhds = boston.groupby('neighborhood')
#print len(nbhds)   # used to later determine the number of clusters used
```

```python
# extract the locational data that will be used, i.e. latitude and longitude
lon = boston.LONGITUDE
lat = boston.LATITUDE
#print len(lats[0:10])

# collect only the values from the series
lon = lon.values   # X-axis
lat = lat.values   # Y-axis

# create StandardScaler object then fit and transform series
scaler = SS()    # default collects mean and std as well
scl_lon = scaler.fit_transform(lon)
scl_lat = scaler.fit_transform(lat)

# free up vm
del scaler
del lon,lat

# create a base plot
plt.scatter(scl_lon,scl_lat)
```

```python
pos = np.array([scl_lon,scl_lat])
pos = np.swapaxes(pos,1,0)
print pos.shape

plt.figure(1)
sns.heatmap(pos[0:100,0:100], xticklabels=False, yticklabels=False, linewidths=0,cbar=False)

edist = metrics.euclidean_distances(pos[0:100,0:100])
plt.figure(2)
sns.heatmap(edist, xticklabels=False, yticklabels=False, linewidths=0, square=True, cbar=False)
```

```python
# error function
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_
        print error[k]

    plt.figure(1)
    plt.plot(range(1,len(error)),error[1:],)
    plt.title='Error Cluster Eval'
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
    
def sc_evaluate_clusters(X,max_clusters):
    s = np.zeros(max_clusters+1)
    s[0] = 0;
    s[1] = 0;
    for k in range(2,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        s[k] = metrics.silhouette_score(X,kmeans.labels_,metric='cosine')
    plt.figure(2)
    plt.plot(range(2,len(s)),s[2:])
    plt.title='SC Cluster Eval'
    plt.xlabel('Number of clusters')
    plt.ylabel('Adjusted Rand Index')
```

```python
# set the clustering of all of the data to the number of neighborhoods - seems natural choice to me
k1 = len(nbhds) # cluster by neighborhoods
kmeans1 = KMeans(init='k-means++',n_clusters=k1,n_init=10)
kmeans1.fit_predict(pos)
#cntrs1 = kmeans1.cluster_centers_
labels1 = kmeans1.labels_
error1 = kmeans1.inertia_
print "The total error of the clustering is: ", error1

evaluate_clusters(pos,k1)   # through plotting errors
```

```python
# from error chart, it looks as if there is a 'knee' at k=4
# try a new clustering scheme from k=4
k2 = 4
kmeans2 = KMeans(init='k-means++',n_clusters=k2,n_init=10)
kmeans2.fit_predict(pos)
cntrs2 = kmeans2.cluster_centers_
labels2 = kmeans2.labels_
error2 = kmeans2.inertia_
print "The total error of the clustering is: ", error2

evaluate_clusters(pos,k2)   # through plotting errors
```

```python
# additionally, try another clustering, k=6
k3 = 6
kmeans3 = KMeans(init='k-means++',n_clusters=k3,n_init=10)
kmeans3.fit_predict(pos)
cntrs3 = kmeans3.cluster_centers_
labels3 = kmeans3.labels_
error3 = kmeans3.inertia_
print "The total error of the clustering is: ", error3

evaluate_clusters(pos,k3)   # through plotting errors
```

```python
# collect the subplots, set them to the same x-axis length
# note that pltarr contains two plots
fig, pltarr = plt.subplots(3,sharex=True,figsize=(15,15))

# set and extend the color maps to accommodate a different color for each neighborhood
cmap = sns.color_palette('deep', n_colors=10)
cmap.extend(sns.color_palette('dark',n_colors=10))
cmap.extend(sns.color_palette('pastel',n_colors=3))

# for each label, isolate the points of that label and plot them with the same color, for k = 23
# uses kmeans1, labels1
for i, label in enumerate(set(labels1)):
    point_indices = np.where(labels1 == label)[0]
    point_indices = point_indices.tolist()
    pltarr[0].scatter(scl_lon[point_indices], scl_lat[point_indices], c=cmap[i], s=20, alpha=0.2, marker='+')
    pltarr[0].set_title('Clustering by Boston Neighborhoods (K = 23)')

# set a new color map for the new clustering, k=4
cmap = sns.color_palette('muted',n_colors=k2)
    
# for each label, isolate the points of that label and plot them with the same color, for k = 4
# uses kmeans2, labels2
for i, label in enumerate(set(labels2)):
    point_indices = np.where(labels2 == label)[0]
    point_indices = point_indices.tolist()
    pltarr[1].scatter(scl_lon[point_indices], scl_lat[point_indices], c=cmap[i], s=20, alpha=0.2, marker='+')
    pltarr[1].set_title('Alternative Clustering (K = 4)')    
    
# set a new color map for the final clustering, k=6
cmap = sns.color_palette('bright',n_colors=k3)

# for each label, isolate the points of that label and plot them with the same color, for k = 6
# uses kmean3, labels3
for i, label in enumerate(set(labels3)):
    point_indices = np.where(labels3 == label)[0]
    point_indices = point_indices.tolist()
    pltarr[2].scatter(scl_lon[point_indices], scl_lat[point_indices], c=cmap[i], s=20, alpha=0.2, marker='+')
    pltarr[2].set_title('Optimized Clustering (K = 6) - Final Plot for Problem 1')
    
```

```python
boston.OPEN_DT = pd.to_datetime(boston.OPEN_DT) # collect the dates of when cases were opened

days = []
for i in range(len(boston.OPEN_DT)):
    date = boston.OPEN_DT[i]
    mn = date.month
    
    if mn == 1: days.append(date.day)   # if the month if February, add 31 to make a continuous list of numbers
    else: days.append(date.day + 31)

days = pd.Series(days, name='Days') # convert into series

ids = boston.CASE_ENQUIRY_ID

# scale both series
scaler = SS()
scl_days = scaler.fit_transform(days)
scl_ids = scaler.fit_transform(ids)
```

```python
arr = np.array([scl_days,scl_ids]) # create an array to input into the kmeans algorithm
arr = np.swapaxes(arr,1,0) # swap to get the right axes set
print arr.shape

# run just to see the error curve
k=10
kmeans = KMeans(init='k-means++',n_clusters=k,n_init=10)
kmeans.fit_predict(arr)
#cntrs1 = kmeans.cluster_centers_
#labels = kmeans.labels_
error = kmeans.inertia_
print "The total error of the clustering is: ", error

evaluate_clusters(arr, k)
```

```python
# after looking at curve and at the values, set k = 4 as optimal clustering amount
k=4
kmeans = KMeans(init='k-means++',n_clusters=k,n_init=10)
kmeans.fit_predict(arr)
cntrs1 = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_
print "The total error of the clustering is: ", error
```

```python
# collect the subplots, set them to the same x-axis length
# note that pltarr contains two plots
fig, pltarr = plt.subplots(1,sharex=True,figsize=(15,15))

# set and extend the color maps to accommodate a different color for each neighborhood
cmap = sns.color_palette('deep', n_colors=k)

# Note that this plot will be CASE_ENQUIRY_IDS (y-axis) vs. Days of Operation (x-axis)
for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    pltarr.scatter(scl_days[point_indices], scl_ids[point_indices], c=cmap[i], s=20, alpha=0.2, marker='+')
    pltarr.set_title('Clustering of Days of Operation with Case Enquiry IDs (K = 4)')
```
