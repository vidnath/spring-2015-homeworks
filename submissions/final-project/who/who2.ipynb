## Hypothesis 1.0: Do higher literacy rates translate to lower rates of HIV/AIDS incidences?

#### First, collect the relevant variables from all of the countries.

```python
# collect literacy rates
lit_total = who['Literacy_rate_adult_total'] # % of total population

# collect HIV/AIDS rates
hiv = who['HIV_infected'] # % of total population
```

```python
# collect the median values to account for the values which are nonexistant
lit_total_med = lit_total.median()
hiv_med = hiv.median()

print lit_total_med, hiv_med
```

```python
# fill in the empty values
lit_total = lit_total.fillna(lit_total_med)
hiv = hiv.fillna(hiv_med)
```

```python
# collect a scatter plot of hiv incidence rates vs total adult literacy rates
plt.scatter(lit_total,hiv)
print lit_total
print lit_total.ndim
print hiv.ndim
```

```python
# clearly doesn't show a lot of information. Let's see what a linear regression says about it
model = sm.OLS(hiv, lit_total)
results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params
```

```python
# add in the line of best fit
# the sm model wasn't working so find another way
# source: http://stackoverflow.com/questions/6148207/linear-regression-with-matplotlib-numpy

fit = pylab.polyfit(lit_total,hiv,1)
fit_fn = pylab.poly1d(fit) # fit_fn is now a function which takes in x and returns an estimate for y
plt.plot(lit_total,hiv, 'b.', lit_total, fit_fn(lit_total), '--k')
plt.xlabel("Literacy Rate (% of Population)")
plt.ylabel("HIV Infected (% of Population)")
plt.title('HIV Prevalence vs. Literacy Rates')
plt.annotate('Mozambique', xy=(lit_total[122]+0.5, hiv[122]+0.5), xytext=(40, 15),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.savefig('hiv_vs_lit.png')
plt.show()
```

```python
# error function
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_
        print error[k]

    plt.figure(1)
    plt.plot(range(1,len(error)),error[1:],)
    plt.title='Error Cluster Eval'
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')
    
def sc_evaluate_clusters(X,max_clusters):
    s = np.zeros(max_clusters+1)
    s[0] = 0;
    s[1] = 0;
    for k in range(2,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        s[k] = metrics.silhouette_score(X,kmeans.labels_,metric='cosine')
    plt.figure(2)
    plt.plot(range(2,len(s)),s[2:])
    plt.title='SC Cluster Eval'
    plt.xlabel('Number of clusters')
    plt.ylabel('Adjusted Rand Index')
```

```python
# set up lists and series to run through and add
country = who.Country
highest_rate = pd.Series(0,index=range(len(country)))
cancer_type = pd.Series('a',index=range(len(country)))
cancers = [colon, liver, lung,stomach]#,hiv,tb_pct]
labels = ['Colon_and_Rectum','Liver','Lung','Stomach']
```

```python
# function to get the maximum value of the different rates of cancer for each country.
def get_max_val(m_c,types,labels,m_p):
    for i in range(len(m_c)):
        rate = types[i]
        
        if rate > m_c[i]:
            m_c[i] = rate 
            m_p[i] = labels
        else:
            continue
```

```python
for i in range(len(cancers)):
    get_max_val(highest_rate,cancers[i],labels[i],cancer_type)
```
