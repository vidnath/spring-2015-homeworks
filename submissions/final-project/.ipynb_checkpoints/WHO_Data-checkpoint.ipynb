```python
# reassing literacy to sorted who dataframe by descending adult literacy rates
lit_desc = who.sort('Literacy_rate_adult_total', ascending=False)
# see what the top 10 countries for adult literacy rates are
lit_desc.head(10)
```

```python
# extract the top 10 countries to get a new dataframe
lit_hi_10 = lit_desc.loc[45:189]
lit_hi_10
```

```python
# get the actual percentage
lit_hi = lit_hi_10['Literacy_rate_adult_total']
lit_hi_countries = lit_hi_10['Country']

lit_lo = lit_lo_10['Literacy_rate_adult_total']
lit_lo_countries = lit_lo_10['Country']
```

```python
# As preventable diseases, first make the assumption that the base population from which the percentages come from is the same for all of
# the statistics. This way, percentages can just be added directly instead of being normalized.

# 'Preventable diseases' here means those which can be stopped by vaccination and education about disease prevention and spread

# Preventable diseases included: HIV/AIDS and tuberculosis. HIV/AIDS is through education, TB is via understanding vaccinations
# Measeles would be included, but metric is only for children, so it will be a biased percentage
```

```python
# In order to get the correct display, percentage from the variables below needs to be multiplied by the population
hiv_hi = lit_hi_10['Prevalence of HIV among adults aged &gt;=15 years (per 100 000 population)'] / 100000
hiv_lo = lit_lo_10['Prevalence of HIV among adults aged &gt;=15 years (per 100 000 population)'] / 100000
tb_hi = lit_hi_10['Incidence of tuberculosis (per 100 000 population per year)'] / 100000
tb_lo = lit_lo_10['Incidence of tuberculosis (per 100 000 population per year)'] / 100000

tb_lo.head(10)

# collect populations
#pop_all = who['Population_total']
#pop_hi = lit_hi_10['Population_total']
#pop_lo = lit_lo_10['Population_total']

# since there is missing data, collect the median values to fill NaN use median to avoid complications from countries that 
# have excessively high or low rates
# first collect counts for the two diseases and divide by 100000 to get percentage
hiv_pct = who['Prevalence of HIV among adults aged &gt;=15 years (per 100 000 population)'] / 100000
tb_pct = who['Incidence of tuberculosis (per 100 000 population per year)'] / 100000
hiv_med = hiv_pct.median()
tb_med = tb_pct.median()

#print hiv_pct.head(10),tb_pct.head(10)

#print hiv_pct.head(10),tb_pct.head(10)

# confirm that NaN values were filled
#print hiv_fill.head(10),tb_fill.head(10)

# fill in the missing hiv and tb data
hiv_hi.fillna(hiv_med)
hiv_lo.fillna(hiv_med)
tb_hi.fillna(tb_med)
tb_lo.fillna(tb_med)

#print hiv_hi
#print hiv_lo
#print tb_lo
#print tb_hi
#print hiv_med

# replace bad value
hiv_lo[58] = hiv_med
#print hiv_lo
```

```python
print min(hiv_lo),min(tb_lo)
print max(hiv_lo),max(tb_lo)
print min(lit_lo),max(lit_lo)
# now plot the values - low values first

l_lo = lit_lo.order(ascending=True)
h_lo = hiv_lo.order(ascending=True)
t_lo = tb_lo.order(ascending=True)


plt.plot(l_lo,h_lo,label='HIV/AIDS Incidence Rate')
plt.plot(l_lo,t_lo,label='Tuberculosis Incidence Rate')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Literacy Rate (%)')
plt.ylabel('HIV/AIDS and TB Incidence Rates (%)')
plt.title('HIV/AIDS and Tuberculosis Incidence rates in Top 10 Least Literate Countries')

# re-scale axis
xmin = min(lit_lo) - 0.01*min(lit_lo)
xmax = max(lit_lo) + 0.01*max(lit_lo)
ymin = 0
ymax = max(hiv_lo) + 0.1*max(hiv_lo)
plt.axis([xmin,xmax,ymin,ymax])
plt.savefig('lit_rates_lo.png')
plt.show()
#plt.axis([22,40,min(lit_tb_lo),max(hiv_hi)])
```

```python
# Second Hypothesis: Do more physicians correlate to lower mortality rates?
# Steps:
# 1. Collect number of physicans
# 2. Collect mortality rates as number of population
# 3. Compare
```

```python
pop_all = who['Population_total']
num_phys = who['Number of physicians'] # number of physicians
adult_mort_pct = who['Adult mortality rate (probability of dying between 15 to 60 years per 1000 population) both sexes'] / 1000

#print adult_mort_pct.head(10)

# fill NaN values
pop_med = pop_all.median()
pop_all = pop_all.fillna(pop_med)   # fill total population counts

phys_med = num_phys.median()
num_phys = num_phys.fillna(phys_med)  # fill physician counts

mort_pct_med = adult_mort_pct.median()
adult_mort_pct = adult_mort_pct.fillna(mort_pct_med)   # fill mortatility percentage


# collect physician percentage
phys_pct = num_phys / pop_all
print phys_pct.head(10)

mort_asc = adult_mort_pct.order(ascending=True)
phys_asc = phys_pct.order(ascending=True)
plt.plot(phys_asc,mort_asc)
plt.xlabel("Number of Physicians (% of Total Population)")
plt.ylabel('Mortality Rate')
plt.title('Mortality Rate vs. Percentage of Physicians')
plt.savefig('mort_phys_plot.png')
```

```python
# Collect the 10 countries with highest and lowest percentage of doctors and their mortality rates
phys_pct = num_phys / pop_all
phys_pct *= 100
print phys_pct.head(10)

# In order to get the mortality rates associated with doctor percentages, a new dataframe needs to be created which has an additionaly
# column for the doctor percentages
who2 = who
who2['Doctor_percentages'] = phys_pct
who2.info() # now has 359 columns instead of 358

# sort into descending order for top 10 highest doctor percentages
phys_desc = phys_pct.order(ascending=False)
phys_hi_10 = phys_desc.loc[149:45]

# sor into ascending order for bottom 10 doctor percentages
phys_asc = phys_pct.order(ascending=True)
phys_lo_10 = phys_asc.loc[134:153]
```
